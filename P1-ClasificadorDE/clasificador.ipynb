# Clasificador por método de Distancia Euclidiana

```python
from sklearn import datasets
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
dataset_cancer= datasets.load_breast_cancer()
dataset_diabetes= datasets.load_diabetes()
dataset_iris= datasets.load_iris()
dataset_vino= datasets.load_wine()
```

# Gráficas de los datasets

```python
print("plot de las caracteristicas del conjunto de datos de iris")
colors = np.array(['pink', 'blue', 'yellow'])
_, ax = plt.subplots()
scatter = ax.scatter(dataset_iris.data[:, 0], dataset_iris.data[:, 1], c=colors[dataset_iris.target])
ax.set(xlabel=dataset_iris.feature_names[0], ylabel=dataset_iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], dataset_iris.target_names, loc="lower right", title="Classes"
)
```

```python
print("PCA de las caracteristicas del conjunto de datos de iris")
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(dataset_iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=dataset_iris.target,
    s=40,
)

ax.set_title("First three PCA dimensions")
ax.set_xlabel("1st Eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd Eigenvector")
ax.zaxis.set_ticklabels([])

plt.show()
```

```python
print("plot de las caracteristicas del conjunto de datos de vino")
_, ax = plt.subplots()
scatter = ax.scatter(dataset_vino.data[:, 0], dataset_vino.data[:, 1], c=dataset_vino.target)
ax.set(xlabel=dataset_vino.feature_names[0], ylabel=dataset_vino.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], dataset_vino.target_names, loc="lower right", title="Classes"
)
```

```python
# Nombres de las características
feature_names = dataset_diabetes.feature_names

# Crear una figura y un conjunto de subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
fig.suptitle('Características del conjunto de datos de diabetes')

# Graficar las características (primeras 9 características para una visualización más clara)
for i, ax in enumerate(axes.flatten()):
    if i < len(feature_names):
        ax.scatter(dataset_diabetes.data[:, i], dataset_diabetes.target)
        ax.set_xlabel(feature_names[i])
        ax.set_ylabel('Progressión de la diabetes')
    else:
        ax.axis('off')

plt.tight_layout()
plt.subplots_adjust(top=0.95)  # Ajustar el espacio para el título
plt.show()
```
